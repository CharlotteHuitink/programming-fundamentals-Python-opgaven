## Deel 4: Basis Operaties

### Wiskundige operaties

````python
print("10 + 3 =", 10 + 3)      # Optellen
print("10 - 3 =", 10 - 3)      # Aftrekken
print("10 * 3 =", 10 * 3)      # Vermenigvuldigen
print("10 / 3 =", 10 / 3)      # Delen
print("10 // 3 =", 10 // 3)    # Integer division (geen decimalen)
print("10 % 3 =", 10 % 3)      # Modulo (remainder) - wat overblijft na integer division
print("10 ** 3 =", 10 ** 3)    # Exponent (machten)
````

### Vergelijkingen
Deze geven `True` of `False` terug

````python
print("5 == 5:", 5 == 5)       # Gelijk aan
print("5 != 3:", 5 != 3)       # Niet gelijk aan
print("5 > 3:", 5 > 3)         # Groter dan
print("5 < 3:", 5 < 3)         # Kleiner dan
print("5 >= 5:", 5 >= 5)       # Groter dan of gelijk aan
print("5 <= 3:", 5 <= 3)       # Kleiner dan of gelijk aan
````

### Logical Operations

````python
print("Waar en onwaar:", True and False)
print("Waar of Onwaar:", True or False)
print("niet Waar:", not True)
````

## Deel 5: Lists - Collecties van Items (15 min)

*Lists* zijn geordende collecties - zoals een boodschappenlijstje of een to-do lijstje.

**Dit is √©√©n van de meest belangrijke datastructuren in Python!** üåü

````python
# Het maken van lists
fruit = ["appel", "banaan", "sinaasappel", "druif"]
getallen = [10, 20, 30, 40, 50]
gemixt = [1, "hallo", 3.14, True]  # Lists kunnen verschillende datatypes bevatten!

print("Fruit:", fruit)
print("Getallen:", getallen)
print("Gemixt:", gemixt)
````

### Toegang tot List Items
‚ö†Ô∏è **Belangrijk:** In Python start het indexeren bij 0!

````python
print("Eerste vrucht:", fruit[0])    # Eerste item (index 0)
print("Tweede vrucht:", fruit[1])   # Tweede item (index 1)
print("Laatste vrucht:", fruit[-1])    # Laatste item
print("Voorlaatste:", fruit[-2])
````

### Slicing - Het verkrijgen van meerdere elementen uit een lijst
Met blokhaken kun je elementen selecteren.

````python
print("Eerste twee getallen:", getallen[0:2])    # Items 0 en 1 (exclusief 2)
print("Vanaf het eerste getal:", getallen[1:])   # Vanaf index 1 tot het eind
print("De eerste drie getallen:", getallen[:3])  # Vanaf het begin tot index 3
````

### List Operaties
Je kunt ook allerlei handelingen uitvoeren op een lijst. Zoals elementen toevoegen, of juist verwijderen. De onderstaande voorbeelden laten zien hoe je dat doet.

````python
fruit = ["appel", "banaan", "sinaasappel"]

print("Originele list:", fruit)
print("Length:", len(fruit))

# Adding items
fruit.append("mango")  # Voeg element toe aan het einde
print("Na 'append':", fruit)

fruit.insert(1, "kiwi")  # Invoegen op positie 1
print("Na het invoegen:", fruit)
````

````python
# Het verwijderen van items
fruit = ["appel", "banaan", "sinaasappel"]
print(fruit)
fruit.remove("banaan")  # Verwijder een specifiek item
print("Na 'remove':", fruit)

laatste_fruit = fruit.pop()  # Verwijder en geef laatste item terug
print("Popped:", laatste_fruit)
print("Na 'pop':", fruit)
````

````python
# Handige list functies
getallen = [3, 1, 4, 1, 5, 9, 2, 6]

print("Getallen:", getallen)
print("Gesorteerd:", sorted(getallen))
print("Max:", max(getallen))
print("Min:", min(getallen))
print("Som:", sum(getallen))
````

## Deel 6: Dictionaries - Key-Value Pairs

*Dictionaries* slaan informatie op in **key-value pairs**, net als een echt woordenboek waarin je een woord (key) opzoekt om de definitie (value) ervan te vinden.

````python
# Het maken van een dictionary
student = {
    "naam": "naam",
    "leeftijd": 20,
    "leerlijn": "Data Science",
    "cijfer": 8.3
}

print("Student dictionary:", student)
````

````python
# Toegang tot de values via de key
print("Naam:", student["naam"])
print("Leeftijd:", student["leeftijd"])
print("Leerlijn:", student["leerlijn"])
````

````python
# Toevoegen/wijzigen
student["email"] = "alice@university.edu"  # Toevoegen van een nieuw key-value pair
student["leeftijd"] = 21  # Het veranderen van een bestaande value

print("Na de update:", student)
````

````python
# Dictionary methods
print("Keys:", list(student.keys()))
print("Values:", list(student.values()))
print("Items:", list(student.items()))
print("\nIs 'naam' in dictionary?", "Data Science" in student)
print("Is 'telefoonnummer' in dictionary?", "telefoonnummer" in student)
````

````python
values_student = list(student.values())
print("Data Science" in values_student)
````

## Part 7: Control Flow - If/Else 
Met *Control Flow* kan je code maken dat zelf beslissingen neemt!

### If Statements

````python
leeftijd = 12

if leeftijd >= 18:
    print("Je bent een volwassene")
else:
    print("Je bent niet volwassen")
````

### If-Else

````python
temperatuur = 35

if temperatuur > 30:
    print("Het is heet!")
else:
    print("Het is niet te heet")
````

### If-Elif-Else

````python
score = 85

if score >= 90:
    waardering = "uitmuntend"
elif score >= 80:
    waardering = "zeer goed"
elif score >= 70:
    waardering = "goed"
elif score >= 60:
    waardering = "voldoende"
else:
    waardering = "onvoldoende"

print(f"Score: {score}, Waardering: {waardering}")
````

### Multiple Conditions

````python
leeftijd = 15
heeft_rijbewijs = False

if leeftijd >= 18 and heeft_rijbewijs:
    print("Jij kunt rijden!")
elif leeftijd >= 18 and not heeft_rijbewijs:
    print("Je bent oud genoeg om een rijbewijs te halen")
else:
    print("Je bent te jong om te rijden")
````

## Part 8: Loops - Acties Herhalen

Met loops kun je code meerdere keren herhalen zonder dat je deze steeds opnieuw hoeft te schrijven.

### For Loops

````python
# Tot 5 tellen
print("Tellen:")
for i in range(1, 6):  # range(1, 6) geeft 1, 2, 3, 4, 5
    print(i)
````

````python
# Door een list loopen
fruit = ["appel", "banaan", "sinaasappel"]

print("Ik houd van deze vrucht:")
for vrucht in fruit:
    print(f"  - {vrucht}")
````

````python
# Looping met een index
print("Numbered list:")
for i, fruit in enumerate(fruit):
    print(f"{i+1}. {fruit}")
````

### While Loops

````python
# Countdown
count = 5
print("Countdown:")
while count > 0:
    print(count)
    count = count - 1  # Dit is hetzelfde als: count -= 1
print("Blast off! üöÄ")
````

## Part 9: Functies - Herbruikbare Code

Functies zijn herbruikbare codeblokken. Schrijf ze √©√©n keer en gebruik ze meerdere keren! ‚ôªÔ∏è

````python
# Een eenvoudige functie defini√´ren
def python_groet():
    print("Hello!")
    print("Welcome to Python!")

# Functie aanroepen
python_groet()
````

### Functies met Parameters

````python
def groet_persoon(naam):
    print(f"Hallo, {naam}!")
    print(f"Leuk om je te ontmoeten, {naam}!")

groet_persoon("Alice")
groet_persoon("Bob")
````

````python
# Meerdere parameters
def introduce(naam, leeftijd):
    print(f"Mijn naam is {naam} en ik ben {leeftijd} jaar oud")

introduce("Jan", 25)
````

### Functies die een waarde teruggeven

````python
def add_numbers(a, b):
    result = a + b
    return result

sum1 = add_numbers(5, 3)
sum2 = add_numbers(10, 20)

print(f"5 + 3 = {sum1}")
print(f"10 + 20 = {sum2}")
````

### Practisch Voorbeeld

````python
# Temperatuur converteren
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}¬∞C is {temp_f}¬∞F")

print(f"0¬∞C is {celsius_to_fahrenheit(0)}¬∞F")
print(f"100¬∞C is {celsius_to_fahrenheit(100)}¬∞F")
````

## üéì Mini Project: Studenten Cijfer Calculator

Laten we alles combineren wat je tot nu toe hebt geleerd!

````python
# Student data
studenten = [
    {"naam": "Alice", "cijfers": [7.5, 9.0, 8.8]},
    {"naam": "Piet", "cijfers": [6.3, 7.2, 7.5]},
    {"naam": "Jan", "cijfers": [3.2, 5.5, 6.0]},
    {"naam": "Diana", "cijfers": [7.3, 8.6, 8.1]}
]

# Functie om het gemiddelde te berekenen
def bereken_gemiddelde(cijfers):
    totaal = sum(cijfers)
    gemiddelde = totaal / len(cijfers)
    return gemiddelde

# Functie om het cijfer te bepalen
def get_cijfer(average):
    if gemiddelde >= 90:
        return "Uitmuntend"
    elif gemiddelde >= 80:
        return "Zeer goed"
    elif gemiddelde >= 70:
        return "Goed"
    elif gemiddelde >= 60:
        return "Voldoende"
    else:
        return "Onvoldoende"

# Doe dit voor alle studenten
print("Student Raport:")
print("-" * 40)

for student in studenten:
    naam = student["naam"]
    cijfers = student["cijfers"]
    gemiddelde = bereken_gemiddelde(cijfers)
    cijfer = get_cijfer(gemiddelde)
    
    print(f"{naam}:")
    print(f"  Cijfers: {cijfers}")
    print(f"  Gemiddelde: {gemiddelde:.1f}")
    print(f"  Cijfer: {cijfer}")
    print()
````
